```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
print("Done")
```

```python
(x_train,y_train), (x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train_oneD = x_train.reshape(len(x_train), 28*28)
x_test_oneD = x_test.reshape(len(x_test),28*28)
```

```python
x_train_oneD[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train_oneD, y_train, epochs=5)
```

```python
model.evaluate(x_test_oneD, y_test)
```

```python
preds = model.predict(x_test_oneD)
preds[0]
```

```python
plt.matshow(x_test[0])
```

```python
preds_lables = [np.argmax(i) for i in preds]
preds_lables[:5]
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions = preds_lables)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10,  activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train_oneD, y_train, epochs=5)
```

```python
model.evaluate(x_test_oneD,y_test)
```

```python
preds = model.predict(x_test_oneD)
preds[0]
```

```python
pred_lables = [np.argmax(i) for i in preds]
pred_lables[:5]
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions=pred_lables)
cm
```

```python
plt.figure(figsize = (10,9))
sn.heatmap(cm,annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
def test_optimizer(x):
    model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10,  activation='sigmoid')
])

    model.compile(optimizer=x,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

    model.fit(x_train, y_train, epochs=5, verbose=0)
    acc = model.evaluate(x_test,y_test, verbose=0)
    return acc
```

```python
optimizer = ['Adadelta','Adagrad', 'Adam', 'Adamax', 'FTRL', 'NAdam', 'RMSprop']
accuracy= {}
for i in optimizer:
    temp = test_optimizer(i)
    accuracy[i] = temp[1]
    print("Accuracy for Optimizer:",i,"->",temp[1])
# accuracy
```

**Got best accurac with "NAdam"**

```python
def test_loss(x):
    model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10,  activation='sigmoid')
])

    model.compile(optimizer='NAdam',
              loss=x,
              metrics=['accuracy'])

    model.fit(x_train, y_train, epochs=5, verbose=0)
    acc = model.evaluate(x_test,y_test, verbose=0)
    return acc
```

```python
loss = ['sparse_categorical_crossentropy','hinge', 'mean_squared_error',
        'huber_loss', 'kl_divergence','log_cosh']
# loss = ['sparse_categorical_crossentropy','categorical_crossentropy','cosine_similarity', 'hinge', 'mean_squared_error'
#         'huber_loss', 'kl_divergence','log_cosh','binary_crossentropy']
accuracy= {}
for i in loss:
    temp = test_loss(i)
    accuracy[i] = temp[1]
    print("Accuracy for Loss:",i,"->",temp[1])
accuracy
```
